Создание веб-приложения, с помощью которого исследователь будет поощрять участников эксперимента чашкой кофе в университетской кофейне.

Контекст задачи: вы проводите лингвистические онлайн-эксперименты и хотите повысить вовлеченность участников, предлагая им бесплатный кофе в университетской кофейне после завершения опроса.

Разработайте веб-приложение на Flask с тремя интерфейсами для разных типов пользователей. Весь проект должен быть выложен на гитхаб.

#### Требования к функциональности

1. Интерфейс для участников эксперимента
- Форма регистрации с полями: имя, фамилия, возраст, программа+курс или должность.
- Анкета с как минимум двумя вопросами на лингвистическую тематику.
- Результат: после отправки формы (данные и ответы записываются в БД) участник получает уникальный код для получения кофе (код тоже записывается в БД).

2. Интерфейс для бариста
- Поле для ввода кода от участника.
- Проверка валидности: система показывает, действителен ли код и не был ли он уже использован.
- Можно "выдать" кофе по кнопке, тогда в БД этот код помечается как использованный.

3. Интерфейс для исследователя-администратора со статистикой

- Общее количество участников.
- Количество выданных кофе.
- Количество неиспользованных кодов.

#### Структура базы данных (SQLite)

Таблица participants должна содержать:
- id (primary key)
- name 
- surname
- email
- age 
- occupation
- code (уникальный код)
- is_used (флаг использования кода)
- created_at (дата и время регистрации)
- used_at (дата и время получения кофе)
- answer_id (foreign key)

Таблица answers должна содержать:
- id (primary key)
- survey_answers (ответы на анкету в формате JSON)

#### Маршруты Flask:
GET / - главная страница с навигацией

GET|POST /participant - форма для участников

GET|POST /barista - интерфейс бариста

GET /admin - панель исследователя

#### Критерии оценивания
1. Программа запускается без дополнительной модификации. (1 балл)
2. Есть база данных с описанной ниже схемой. (1 балл)
3. Есть (главная) страница для участников опроса. На странице есть анкета, описанная выше. (1 балл)
4. При отправке анкеты сведения из неё записываются в базу данных. (1 балл)
5. При отправке анкеты генерируется и записывается в базу данных код для получения кофе. Есть страница для бариста. На странице есть интерфейс для ввода и проверки кода. (1 балл)
6. Страница для бариста работает: на ней можно и проверить, и погасить код для получения кофе. (1 балл)
7. Есть страница для исследователя-администратора. (1 балл)
8. Страница для исследователя-администратора работает: на основании данных в базе данных выводится количество участников, выданных кофе и неиспользованных кодов. (1 балл)
